# Supermart Grocery Sales - Retail Analytics Dataset
 
Dataset downloaded from https://www.kaggle.com/datasets/mohamedharris/supermart-grocery-sales-retail-analytics-dataset

# Import 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc = {'figure.figsize':(15,8)})
sns.set_style('darkgrid')
sns.set_palette('winter')
```

```python
df = pd.read_csv('Sales.csv')
print(df.info())
df.head()
```

# Clean

```python
# Change date column to a date datatype
df['Order Date'] = df['Order Date'].astype('datetime64')
df.info()
```

```python
# Split date column
df['day'] = pd.DatetimeIndex(df['Order Date']).day
df['day_of_the_week'] = pd.DatetimeIndex(df['Order Date']).weekday
df['month'] = pd.DatetimeIndex(df['Order Date']).month
df['year'] = pd.DatetimeIndex(df['Order Date']).year
df
```

# Basic EDA

```python
# Here is a function for basic exploratory data analysis:
def eda(dataframe):
    # Replace any blank spaces w/ a underscore.
    dataframe.columns = dataframe.columns.str.replace(" ", "_")
    # Checks for the null values.
    print("Missing values \n{}".format(dataframe.isnull().sum()))
    print("Total missing values  {}".format(dataframe.isnull().sum().sum())) 
    # Checks the data frame range size.
    print("\nData index {}".format(dataframe.index))
    # Checks for data types of the columns within the data frame.
    print("\nData types \n{}".format(dataframe.dtypes))
    # Checks the shape of the data frame.
    print("\nShape {}".format(dataframe.shape))
    # Gives us any statistical information of the data frame.
    print("\nDescribe \n{}".format(dataframe.describe()))
    # Gives us the duplicated data of the data frame. 
    print("\nDuplicates \n{}".format(dataframe[dataframe.duplicated()].sum()))
    # A for loop that does this for every single column & their 
    # values within our data frame giving us all unique values.
    print('\nUnique Values Count')
    for item in dataframe:
        print(item, dataframe[item].nunique())
    print('\nUnique Values')   
    for item in dataframe:
        print(item,'\n',dataframe[item].unique())
       
# Let's apply this function to our entire data frame.
eda(df)
sns.heatmap(df.isnull(),yticklabels=False,cmap='cividis')
sns.pairplot(df, kind='reg',diag_kind='kde')
```

```python
# Pearson Correlation Heatmap
sns.heatmap(df.corr(method='pearson'),linewidths=.5,cmap="YlGnBu",annot=True)
```

```python
# Spearman Correlation Heatmap
sns.heatmap(df.corr(method='spearman'),linewidths=.5,cmap="YlGnBu",annot=True)
```

```python
# largest positive correlated columns to our tartget or y column.
largest_pos_corr_list = df.corr()[['Profit']].sort_values('Profit').tail(5).T.columns.drop('Profit').tolist()
print('largest_pos_corr_list',largest_pos_corr_list)
# largest negative correlated columns to our tartget or y column.
largest_neg_corr_list = df.corr()[['Profit']].sort_values('Profit').head(5).T.columns.tolist()
print('largest_neg_corr_list',largest_neg_corr_list)
```

# Answering Questions about our data

```python
# Who are our most profitable customers
df.groupby('Customer_Name')['Profit'].sum().sort_values(ascending=False).head()
```

```python
# What is our most poriftable years
df.groupby('year')['Profit'].sum().sort_values(ascending=False)
```

```python
# On average what are our most profitable months
df.groupby('month')['Profit'].mean().sort_values(ascending=False)
```

```python
# What are our most profitable catefories
df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
```

```python
# On average whats is the most profitable category in our most porfitable month
df.groupby(['month','Category'])['Profit'].mean().sort_values(ascending=False)
```

```python
# What are our best customers favorite categories
df.groupby(['Customer_Name','Category'])['Profit'].sum().sort_values(ascending=False).head()
```
