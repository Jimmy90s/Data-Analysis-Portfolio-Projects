```python
import pandas as pd 
import seaborn as sns
import sqlalchemy
sns.set(rc = {'figure.figsize':(15,8)})
conn = sqlalchemy.create_engine('mssql+pyodbc://localhost/master?driver=SQL+Server+Native+Client+11.0')
%load_ext sql
%sql mssql+pyodbc://localhost/master?driver=SQL+Server+Native+Client+11.0
```

```python
 bg = pd.read_sql_query('''

SELECT TOP 4 Region , SUM(Profit) AS [Total Profit]
FROM [master].[dbo].[Sales]
GROUP BY Region
ORDER BY SUM(Profit) DESC

''', conn)
sns.barplot(data=bg,x='Total Profit',y='Region').set(title='Profit by Region')
```

```python
 bh = pd.read_sql_query('''

SELECT Category , SUM(Profit) AS [Total Profit]
FROM [master].[dbo].[Sales]
GROUP BY Category
ORDER BY SUM(Profit) DESC

''', conn)
sns.barplot(data=bh,x='Total Profit',y='Category').set(title='Profit by Category')
```

```python
bb = pd.read_sql_query('''

SELECT Profit
FROM [master].[dbo].[Sales]

''', conn)
sns.boxplot(data=bb)

```
